```python
import pandas as pd
import joblib
from datasets import load_dataset
```

```python
dataset = load_dataset("aai510-group1/telco-customer-churn")
df = dataset["train"].to_pandas()
print(df.shape)
df.head()
```

```python
df.to_csv("churn.csv", index=False)
    
```

```python
def get_data(data_csv):
    return pd.read_csv(data_csv)
    
```

```python
def drop_vars(df):
    df = df.drop(["Churn Category", "Churn Reason", "Churn Score", "City", "Country", "Customer ID",
              "Customer Status", "Gender", "Lat Long", "Latitude", "Longitude", "Offer",
              "Quarter", "State", "Zip Code"
             ], axis="columns")
    return df
    
```

```python
def get_dummies(df):
    return pd.get_dummies(df, columns=["Contract", "Payment Method", "Internet Type"], drop_first=True)
    
```

```python
def get_X(df):
    return df.drop(columns="Churn")
```

```python
def load_model(model):
    return joblib.load(model)
```

```python
def run_model(data_csv, model):
    df = get_data(data_csv)
    df = drop_vars(df)
    df = get_dummies(df)
    X = get_X(df)
    m = load_model(model)
    proba = m.predict_proba(X)
    
    return proba
    
```

```python
output = run_model("churn.csv", "churn_model.joblib")
```
